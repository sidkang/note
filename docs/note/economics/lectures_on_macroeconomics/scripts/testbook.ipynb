# Seaborn

```python
import seaborn as sns
import pandas as pd
```

```python
flights_data = sns.load_dataset("flights")
flights_data.head()
```

```python
sns.scatterplot(data=flights_data, x="year", y="passengers")
```

```python
sns.lineplot(data=flights_data.loc[:30], x="year", y="passengers")
```

```python
flights_data.head()
```

```python
sns.set_style("darkgrid")
sns.lineplot(data = flights_data, x = "year", y = "passengers", hue=None)
```

```python
sns.barplot(data=flights_data, x="year", y="passengers")
```

```python
tips = sns.load_dataset("tips")
sns.relplot(x="total_bill", y="tip", hue="size", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="size", palette="ch:r=-.5,l=.75", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", size="size", hue="size", sizes=(15, 200), data=tips)
```
